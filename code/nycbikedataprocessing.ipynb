```python
from datetime import date
# national holiday
us_holidays = holidays.US()
res['US_holiday'] = res[['year', 'month', 'day']].apply(lambda x: (0, 1)[date(x['year'],x['month'],x['day']) in us_holidays], axis = 1)

# new york holiday
ny_holidays = holidays.US(state='NY')
res['NY_holiday'] = res[['year', 'month', 'day']].apply(lambda x: (0, 1)[date(x['year'],x['month'],x['day']) in ny_holidays], axis = 1)
```

```python
# more holidays
# thanksgiving friday
res.loc[(res['year'] == 2013) & (res ['month'] == 11) & (res['day'] == 29), "US_holiday"] = 1
res.loc[(res['year'] == 2014) & (res ['month'] == 11) & (res['day'] == 28), "US_holiday"] = 1
res.loc[(res['year'] == 2015) & (res ['month'] == 11) & (res['day'] == 27), "US_holiday"] = 1

res.loc[(res['year'] == 2013) & (res ['month'] == 11) & (res['day'] == 29), "NY_holiday"] = 1
res.loc[(res['year'] == 2014) & (res ['month'] == 11) & (res['day'] == 28), "NY_holiday"] = 1
res.loc[(res['year'] == 2015) & (res ['month'] == 11) & (res['day'] == 27), "NY_holiday"] = 1
```

```python
# work_day-not a holiday and monday - Friday
res['NY_workday'] = res[['NY_holiday','dayofweek']].apply(lambda x: (0, 1)[x['NY_holiday'] == 0 and x['dayofweek'] <5], axis = 1)
res['US_workday'] = res[['US_holiday','dayofweek']].apply(lambda x: (0, 1)[x['US_holiday'] == 0 and x['dayofweek'] <5], axis = 1)
```
